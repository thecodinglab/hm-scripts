```python
import numpy as np
import dgl_euler as euler

f = lambda x, y: x**2 + 0.1 * y

a = -1.5
b = 1.5
n = 5
y0 = 0

x_classic, y_classic = euler.classic(f, a, b, n, y0, print_steps=True)
x_midpoint, y_midpoint = euler.midpoint(f, a, b, n, y0, print_steps=False)
x_modified, y_modified = euler.modified(f, a, b, n, y0, print_steps=False)
```

```python
import matplotlib.pyplot as plt

x_min = a
x_max = b
y_min = -1
y_max = 3

count = 30
x, y = np.meshgrid(
    np.linspace(x_min, x_max, count),
    np.linspace(y_min, y_max, count)
)

plt.quiver(x, y, 1, f(x, y))

plt.plot(x_classic, y_classic, label="classic")
plt.plot(x_midpoint, y_midpoint, label="midpoint")
plt.plot(x_modified, y_modified, label="modified")
plt.legend()
plt.show()
```
